# Python 基础语法汇总

本 notebook 汇总了常见的 Python 基础语法模块，每个模块包含说明和示例代码。

## 1. 注释、编码声明与输出/输入

```python
# 单行注释

"""
多行注释
"""

# 输出
print("Hello, Python!")        # 打印字符串
print(1, 2, 3, sep=' | ')     # 自定义分隔符
print("不换行", end=' ')       # 不换行，后续输出在同一行
print("\n")                   # 打印空行
```

## 2. 变量与数据类型

```python
# 基本数据类型
a = 10            # 整数 int
b = 3.14          # 浮点数 float
c = True          # 布尔 bool
d = "字符串"       # 字符串 str
e = None          # 空值 NoneType

# 类型检查与转换
print(type(a))           # <class 'int'>
print(float(a), str(a), bool(0))

print(type(b))
print(type(c))
print(type(d))
print(type(e))    
```

## 3. 运算符

```python
# 算术运算
x, y = 7, 3
print(x + y, x - y, x * y, x / y, x // y, x % y, x ** y)

# 赋值运算
x += 1; x -= 2; x *= 3; x /= 4
print(x)

# 比较与逻辑运算
print(x == y, x != y, x > y, x < y)
print(True and False, True or False, not True)

# 成员与身份运算
lst = [1, 2, 3]
print(2 in lst, 5 not in lst)
print(x is y, x is not y)
```

## 4. 字符串常用操作

```python
s = "Hello, World!"
print(s[0], s[7:], s.lower(), s.upper())

print(s.split(", "), s.replace("World", "Python")) # .split 返回list

print("".join(["Py", "thon"]))

name, age = "Alice", 30
print(f"Name: {name}, Age: {age}")
```

## 5. 列表、元组、集合、字典

```python
# 列表 list
lst = [1, 2, 3]
lst.append(4)
lst.pop()
lst[1] = 20
print(lst)

# 元组 tuple
tup = (1, 2, 3)
# tup[0] = 9  # 不可变

# 集合 set
st = {1, 2, 3}
st.add(4)
st.remove(2)
print(st)

# 字典 dict
d = {"a": 1, "b": 2}
d["c"] = 3
print(d.get("d", 0))
del d["b"]
print(d)
```

## 6. 列表推导式 & 生成器表达式

```python
# 列表推导式
squares = [x**2 for x in range(10) if x % 2 == 0]
print(squares)

# 字典推导式
d = {x: x**2 for x in range(5)}
print(d)

# 集合推导式
st = {x**2 for x in range(5)}
print(st)

# 生成器表达式
gen = (x**2 for x in range(10))
for v in gen:
    print(v, end=' ')
```

## 7. 条件与循环

```python
# 条件语句
x = -1
if x > 0:
    print("正数")
elif x == 0:
    print("零")
else:
    print("负数")

# for 循环
for i in range(5):
    print(i, end=' ')

print("\n")                  


# while 循环
i = 0
while i < 5:
    print(i, end=' ')
    i += 1

print("\n")                


# break / continue
for i in range(10):
    if i == 3:
        continue
    if i == 7:
        break
    print(i, end=' ')
```

## 8. 函数与作用域

```python
# 定义函数
def add(a, b=0):
    """
    文档字符串，
    b=0 表示给参数 b 设了一个默认值，如果调用时不传 b，它就自动取 0
    """
    return a + b

print(add(3, 4))

# 可变参数
def func(*args, **kwargs):
    """"
    *args：把所有位置参数（positional arguments）收集到一个元组 args 里。
    **kwargs：把所有关键字参数（keyword arguments）收集到一个字典 kwargs 里。
    """
    print(args, kwargs)

func(1, 2, x=3, y=4)

# 作用域示例
x = 10  # 全局

def foo():
    global x # 在函数体内声明“下面对 x 的赋值，指的是全局作用域里的 x”，从而允许我们修改全局变量。
    x = 20  # 修改全局
    y = 5   # 局部，不带 global，则 y 永远是 foo 的局部变量，函数外无法访问。
    print(x, y)

foo()
print(x)
```

## 9. Lambda 匿名函数 & 常用高阶函数

```python
# lambda
f = lambda x, y: x + y
print(f(2, 3))

# map / filter / reduce
nums = [1, 2, 3, 4]
m = map(lambda x: x*2, nums) # 把这个函数依次作用到 nums 里的每个元素，返回一个可迭代的 map 对象。
print(m)
print(list(map(lambda x: x*2, nums)))
print(list(filter(lambda x: x%2==0, nums)))
```

## 10. 异常处理

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("除零错误：", e)
except Exception as e:
    print("其他错误：", e)
else:
    print("无异常时执行")
finally:
    print("始终执行")
```

## 11. 类与面向对象

```python
class Person:
    """表示一个人的类"""

    # 类属性（所有实例共享）
    species = "Homo sapiens"

    def __init__(self, name: str, age: int):
        # 实例属性（每个实例独有）
        self.name = name
        self.age = age

    def greet(self):
        """实例方法：打招呼"""
        print(f"Hi, 我是 {self.name}，今年 {self.age} 岁。")

# 创建实例
p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

p1.greet()  # Hi, 我是 Alice，今年 30 岁。
p2.greet()  # Hi, 我是 Bob，今年 25 岁。

# 访问类属性
print(Person.species)  # Homo sapiens
print(p1.species)      # Homo sapiens
```
