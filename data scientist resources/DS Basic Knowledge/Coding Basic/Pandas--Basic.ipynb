## 1. 导入 Pandas

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'], name='ExampleSeries')
print(s)

df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': ['x', 'y', 'z'],
    'C': pd.date_range('20250101', periods=3)
})

arr = np.arange(12).reshape(3, 4)

df2 = pd.DataFrame(arr, columns=list('WXYZ'))
print(df1)
print(df2)
```

## 8. 常用函数演示

```python
df_nan = pd.DataFrame({'A': [1, np.nan, 3], 'B': [np.nan, 2, 3]})
df_filled = df_nan.fillna(0)  # fillna()
df_sorted = df2.sort_values(by='Y', ascending=False)  # sort_values()
df_dup = pd.DataFrame({'K': [1, 1, 2, 3], 'V': ['a', 'a', 'b', 'c']})
df_nodup = df_dup.drop_duplicates()  # drop_duplicates()
print(df_nan)
print(df_filled)
print(df_sorted)
print(df_dup)
print(df_nodup)
```

## 9. 统计分析与时间处理

```python
print(df2.describe())
print(df2.mean())
dates = pd.date_range('2025-01-01', periods=6, freq='D')
ds = pd.Series(np.random.randn(6), index=dates)
print(ds)
print(ds.resample('2D').sum())
```

## 10. 分组聚合 (groupby)

```python
df_group = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C','C'],
    'Value': [10, 20, 30, 40, 50, 60]
})

grouped = df_group.groupby('Category')  # groupby()
print(type(grouped))
print(grouped.sum())

def normalize(x):
    return (x - x.mean()) / x.std()

print(grouped['Value'].apply(normalize))  # apply()
```

## 11. 连接与合并 (join)

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'L': [1, 2, 3]})
right = pd.DataFrame({'key': ['K0', 'K2', 'K3'], 'R': [4, 5, 6]})

merged_inner = pd.merge(left, right, on='key', how='inner')
print(merged_inner)  # inner join

merged_outer = pd.merge(left, right, on='key', how='outer')
print(merged_outer)

left2 = left.set_index('key')
right2 = right.set_index('key')
joined = left2.join(right2, how='outer')  # join() 默认是按 索引（index） 来拼接的。

print(joined)
```

## 12. 透视表 (pivot_table)

```python
df_pivot = pd.DataFrame({
    'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250]
})

pivot = pd.pivot_table(df_pivot, 
                       values='Sales', 
                       index='Date', 
                       columns='Category',
                       aggfunc='sum', 
                       fill_value=0)
print(pivot)

# data：原始 DataFrame，指定要透视的源表格。
# values：要聚合的数值列，如 'Sales'。
# index：作为透视表行索引的列名，如 'Date'。
# columns：作为透视表列索引的列名，如 'Category'。
# aggfunc：对同一 (index, columns) 单元格下的值进行汇总的函数，如 sum、mean。
# fill_value：将聚合后结果中的 NaN 替换为指定的填充值。
```
