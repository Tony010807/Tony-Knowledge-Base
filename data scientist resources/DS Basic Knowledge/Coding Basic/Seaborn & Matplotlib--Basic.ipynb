# Matplotlib & Seaborn 基础绘图

本笔记展示如何使用 Matplotlib 和 Seaborn 绘制：
- 柱状图（bar）
- 散点图（scatter）
- 饼图（pie）


```python
# 导入库
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D'],
    'value': [10, 23, 17, 9],
    'x': [1, 2, 3, 4],
    'y': [2.5, 3.8, 1.2, 4.0]
})

df
```

## 1. Matplotlib 柱状图（bar）

```python
plt.figure(figsize=(6,4))
plt.bar(df['category'], df['value'])
plt.title('Matplotlib Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 2. Seaborn 柱状图（barplot）

```python
plt.figure(figsize=(6,4))
sns.barplot(x='category', y='value', data=df)
plt.title('Seaborn Barplot')
plt.show()
```

## 3. Matplotlib 散点图（scatter）

```python
plt.figure(figsize=(6,4))
plt.scatter(df['x'], df['y'])
plt.title('Matplotlib Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 4. Seaborn 散点图（scatterplot）

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='x', y='y', data=df)
plt.title('Seaborn Scatterplot')
plt.show()
```

## 5. Matplotlib 饼图（pie）

```python
plt.figure(figsize=(5,5))
plt.pie(df['x'], labels=df['category'], autopct='%.1f%%')
plt.title('Matplotlib Pie Chart')
plt.axis('equal')  # 保持正圆
plt.show()
```
